```python
class config:
    data = 'Cora' #choices=["Cora", "Citeseer", "Pubmed", "Coauthor-CS"]
    model = 'SimpleGCN' 
    n_hid = 64
    dropout = 0.6
    dropout2 = 0.6
    learning_rate = 0.01
    weight_decay = 5e-4
    patience = 200
    num_iter = 1000
    num_test = 10
    hop = 2
    alpha = 0.1 # used by APPNP only
    random_label_split = False
    num_train = 20 # for random label split only
    num_val = 30 # for random label split only
    data_load = True # load the saved label split to rerun the test (for reproduce purpose)
    test_id = 1 # number of the test, only used to record the ith number of the random label split (for reproduce purpose)
    filter_pct = 0.1 # remove the top and bottom filer_pct points before obtaining statistics of test accuracy
    log = False # whether to show the training log or not
    eval_metric = 'acc' # evaluation metrics, choices=["acc", "f1-macro", "f1-micro"]
    root_dir = '../..' # dir of the source code

args = config()
```

```python
import sys
sys.path.append(args.root_dir)
from train import *
```

```python
accs = []
datasets = ['Cora', 'Citeseer', 'Pubmed', 'Coauthor-CS']
hops = [2, 5, 10, 15, 20, 30, 40, 50]
for dataset in datasets:
    args.data = dataset
    if dataset == 'Cora':
        args.dropout = 0.6
        args.dropout2 = 0.6
        args.weight_decay = 5e-4
        args.learning_rate = 0.01
        print('\nstart testing on ' + dataset + ' dataset')
        acc = []
        for hop in hops:
            args.hop = hop
            test_metrics = main(args)
            test_metrics = remove_edge_pts(test_metrics, pct=args.filter_pct)
            acc.append(test_metrics.mean())
        accs.append(acc)
    elif dataset == 'Citeseer':
        args.dropout = 0.8
        args.dropout2 = 0.6
        args.weight_decay = 5e-4
        args.learning_rate = 0.01
        print('\nstart testing on ' + dataset + ' dataset')
        acc = []
        for hop in hops:
            args.hop = hop
            test_metrics = main(args)
            test_metrics = remove_edge_pts(test_metrics, pct=args.filter_pct)
            acc.append(test_metrics.mean())
        accs.append(acc)
    elif dataset == 'Pubmed':
        args.dropout = 0.8
        args.dropout2 = 0.5
        args.weight_decay = 5e-4
        args.learning_rate = 0.02
        print('\nstart testing on ' + dataset + ' dataset')
        acc = []
        for hop in hops:
            args.hop = hop
            test_metrics = main(args)
            test_metrics = remove_edge_pts(test_metrics, pct=args.filter_pct)
            acc.append(test_metrics.mean())
        accs.append(acc)
    elif dataset == 'Coauthor-CS':
        args.dropout = 0.6
        args.dropout2 = 0.2
        args.weight_decay = 5e-3
        args.learning_rate = 0.01
        print('\nstart testing on ' + dataset + ' dataset with random split: ' + str(2))
        args.test_id = 2
        acc = []
        for hop in hops:
            args.hop = hop
            test_metrics = main(args)
            test_metrics = remove_edge_pts(test_metrics, pct=args.filter_pct)
            acc.append(test_metrics.mean())
        accs.append(acc)
```

```python
accs = np.array(accs)
np.savetxt('acc_test_SimpleGCN_Allhops.txt', accs)
```
